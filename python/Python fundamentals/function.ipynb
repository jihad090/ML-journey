```python
# simple function
def greet():
  print("Hello")
greet()
```

```python
# function with parameter
def sub(a, b):
  return a-b
print(sub(2, 3))

def defaultValuedParameter(b, a=3): # the default valued parameter must come after non valued parameter
  return a-b
print(defaultValuedParameter(1))

def profile(name, age):
  print(f"My name is {name} and now I am {age}")
profile("Jihad", 23)
profile(age=27, name="Fahad") # if we are not sure about the position of parameter, then it is best to explicitely define them
```

```python
# Lambda function (Anonmous Function)
square = lambda x: x*x-x
print(square(5))
```
